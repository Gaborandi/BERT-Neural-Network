```python
#AutoTokenizer : allows you to convert a sequence of texts into numbers (tokens)
#AutoModelForSequenceClassification : it gives you the architicture from the transformers to load into NLP model
#re : it will create a function alllows us to extract the speciefic comments we want from the website 

from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import requests
from bs4 import BeautifulSoup
import re
```

2. Creating Our Model

```python
#instantial model
#we are using a model already pre-trained before, we retrieve this model from this website :
#"https://huggingface.co/nlptown/bert-base-multilingual-uncased-sentiment"
#tokenizer : it is our model that will convert the text to numbers 

tokenizer = AutoTokenizer.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment")
model = AutoModelForSequenceClassification.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment")
```

3. Encode and Calculate Sentiment

```python
#encode and calculate sentiment

tokens = tokenizer.encode("it is ugly, i love it", return_tensors = "pt")
tokens[0] #added [0] is because we can't pass through a list of lists(remove it and try to decode)
```

```python
result = model(tokens)
result
#to understand the output : it is a onehot encoded list of scores, the position with the highest score represents 
#the sentiment rating 
```

```python
#the argument "argmax" gives you the highest score result 
#the output is the position of our sentiment
torch.argmax(result.logits)
```

```python
#we need to convert the position to a score from 1-5 
#you can try different text to analyze it, 1 means the worst and 5 means the best

int(torch.argmax(result.logits))+1
```

4. Collect Reviews

```python
# lets go further and analyze more sophisticated sentiments 
```

```python
#collect reviews 
#thees 5 lines of codes considered as a simple web scraping
 
#r = it uses request package to wrap any website
r = requests.get("https://www.yelp.com/biz/social-brew-cafe-pyrmont")

#here we use the data structure output from r to create a another format enables BeautifulSoup to search within it
soup = BeautifulSoup(r.text, "html.parser")

#regex, results and reviews are used to extract the import comonents from this web page

#regex is to look for any comment within a class (right click then inspect on any comment on the website, 
#if you find a class then followed by a comment,then, this will be extracted by this line of code) 
regex = re.compile('.*comment.*')

#here we are searching for a paragraph "p" , that follow our format (a class followed by comment) from regex code
results = soup.find_all('p',{'class' : regex})

#reviews is to remove all the unwanted taggs and get the needed text only within a list
reviews = [result.text for result in results]
```

5. Load Reviews into DataFrame and Score

```python
#import pandas and numpy to convert our data into structured one 

import pandas as pd
import numpy as np
```

```python
#load our data and score into structured data frame 

df = pd.DataFrame(np.array(reviews), columns = ["reviews"])
df.head()
```

```python
#lets take a row and see it

df["reviews"].iloc[7]
```

```python
#lets create function
#same steps of the previous sentiment analyziz, we used the same code (tokenizer, model and argmax), put replaced
#the text with the website reviews
#the sentiment function allows us the encapsulate the sentiment pipeline in a function makes it easier to process
#multiple strings

def sentiment_analysis (reviews) :
    tokens = tokenizer.encode(reviews, return_tensors = "pt")
    results = model(tokens)
    return int(torch.argmax(results.logits))+1   
```

```python
#lets try our function again, but we only need the score of the review

sentiment_analysis(df["reviews"].iloc[7])
```

```python
#let make the score for all our data
#we will add a column called sentiment analysis includes the scire of our sentiment analysis
#we will create a a loop function (apply)  that go through each row on our reviews col and then give it a score
#x works as each individual review 
#we use our sentiment_analysis function to pass it to x 
#this extra block [512], is due to the fact that NLP pipeline is limited of how much text or tokens you can pass
#through to it at one particular time, which means that will only grap the first 512 tokens from each review

df["sentiment score"] = df["reviews"].apply(lambda x: sentiment_analysis(x[:512]))
```
